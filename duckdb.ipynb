```python
import duckdb
import pandas as pd
import geopandas as gpd
```

```python
# Load duckdb spatial extension
duckdb.sql("INSTALL spatial")
duckdb.sql("LOAD spatial")
counties_path = "data/external/maryland_county_boundaries.geojson"
reports_path = "data/raw/CrashMap_REPORT_data.csv"
nonmotorists_path = "data/raw/CrashMap_NONMOTORIST_data.csv"
counties = gpd.read_file(counties_path)
reports = pd.read_csv(reports_path)
nonmotorists = pd.read_csv(nonmotorists_path)
```

Let's create the counties table

```python
duckdb.sql(
    f"""
    CREATE 
    OR REPLACE TABLE counties AS 
    SELECT 
      county, 
      geom 
    FROM 
      ST_READ('{counties_path}')
    """
)
# Write a query to create a table of reports where the Latitute and Longitude are converted to a geometry column
duckdb.sql(
    f"""
    CREATE 
    OR REPLACE TABLE reports AS 
    SELECT 
    *, 
    ST_POINT(Longitude, Latitude) AS geom 
    FROM 
    '{reports_path}'
"""
)
# Create a table of nonmotorists
duckdb.sql(
    f"""
    CREATE 
    OR REPLACE TABLE nonmotorists AS 
    SELECT 
      * 
    FROM 
      '{nonmotorists_path}'
"""
)
```

```python
# Query those reports that are in the nonmotorists table
duckdb.sql(
    """
    SELECT *
    FROM reports
    WHERE ReportNumber IN (
        SELECT ReportNumber
        FROM nonmotorists
    )
    LIMIT 3
    """
)
```

```python
# Query those reports that are in the nonmotorists table AND are actually located in 'Baltimore City' using a JOIN and ST_WITHIN on the counties table. There is no geometry necessary for the nonmotorists table.
duckdb.sql(
    """
    SELECT *
    FROM reports
    JOIN counties
    ON ST_WITHIN(reports.geom, counties.geom)
    WHERE counties.county = 'Baltimore City'
    AND reports.ReportNumber IN (
        SELECT ReportNumber
        FROM nonmotorists
        )
    """
)
```

```python
selected = duckdb.sql(
    """
    SELECT 
        reports.ReportNumber, 
        reports.geom AS geometry, 
        ST_AsWKB(reports.geom) AS wkb,
        counties.county,
        ST_X(reports.geom) AS longitude,
        ST_Y(reports.geom) AS latitude,
    FROM reports
    JOIN counties
    ON ST_WITHIN(reports.geom, counties.geom)
    WHERE counties.county = 'Baltimore City'
    AND reports.ReportNumber IN (
        SELECT ReportNumber
        FROM nonmotorists
        )
    """
)
selected
```

```python
selected.df()
```

```python
# Create a geodataframe from the selected reports using the longitude and latitude columns

gdf = gpd.GeoDataFrame(
    selected.df(),
    geometry=gpd.points_from_xy(selected.df().longitude, selected.df().latitude),
)
```

```python
import folium
from folium.plugins import HeatMap


# Create a map of the previous query using folium
m = folium.Map(location=[39.2904, -76.6122], zoom_start=12, tiles="cartodbpositron")
heat_data = [[point.xy[1][0], point.xy[0][0]] for point in gdf.geometry]
HeatMap(
    heat_data,
    radius=8,
    blur=6,
    min_opacity=0.3,
    gradient={0.2: "blue", 0.4: "lime", 0.6: "yellow", 1: "red"},
).add_to(m)
m
```
